```python
import pandas as pd
import numpy as np
```

```python
ada = pd.read_csv('ADA_data.csv')
ada
```

```python
columns=['Dividends', 'Stock Splits']
ada.drop(columns, axis=1, inplace=True)
```

```python
ada = ada.assign(Name= 'ADA')
```

```python
ada
```

```python
ada
```

```python
btc = pd.read_csv('Bitcoin_data.csv')
btc
```

```python
columns=['Dividends', 'Stock Splits']
btc.drop(columns, axis=1, inplace=True)
```

```python
btc = btc.assign(Name = 'BTC')
```

```python
btc
```

```python
dgc = pd.read_csv('DodgeCoin_data.csv')
dgc
```

```python
columns=['Dividends', 'Stock Splits']
dgc.drop(columns, axis=1, inplace=True)
```

```python
dgc = dgc.assign(Name= 'DGC')
```

```python
eth = pd.read_csv('Ethereum_data.csv')
eth
```

```python
columns=['Dividends', 'Stock Splits']
eth.drop(columns, axis=1, inplace=True)
```

```python
eth = eth.assign(Name= 'ETH')
```

```python
all_coins = pd.concat([btc, eth, ada, dgc]).reset_index()
all_coins
```

```python
all_coins
```

```python
all_coins.to_csv('all_coins.csv')
```
